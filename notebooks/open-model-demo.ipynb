# DevEval Agents — open-weight model demo (Google Colab)

**Purpose:** reproduce, with one *Runtime → Run all*, that the DevEval Agents
KOICA/ODA evaluation agents run on a **fully open-weight** model (no proprietary
API). This is evidence for the Digital Public Goods Standard, Indicator 4
(Platform Independence).

**Before running:** set a GPU runtime — *Runtime → Change runtime type → T4 GPU*.
The free tier is enough for `qwen2.5:7b`.

It installs [Ollama](https://ollama.com), pulls an open model, clones the repo,
and runs `scripts/open_runner.py` on the bundled sample report.

### 1. Install Ollama and start the local server
(`zstd` is installed first — the Ollama installer needs it to unpack on Colab.)

```python
!apt-get -qq update && apt-get -qq install -y zstd
!curl -fsSL https://ollama.com/install.sh | sh
!nohup ollama serve > /content/ollama.log 2>&1 &
import time; time.sleep(6)
!curl -s http://localhost:11434/api/version || echo 'server starting...'
```

### 2. Pull an open-weight model
`qwen2.5:7b` has open (Apache-2.0) weights and fits the free T4. You can also
try `qwen2.5:14b`, `llama3.1:8b`, `gemma2:9b`, etc. — the point is that the
model is replaceable.

```python
!ollama pull qwen2.5:7b
```

### 3. Clone the project

```python
!git clone --depth 1 https://github.com/amnotyoung/dev-eval-agents.git
```

### 4. Run the same agents on the open model
This injects the KOICA rules + reference knowledge (exactly what the Codex
harness reads from `AGENTS.md` + `reference/`) and evaluates the sample project
report on the OECD-DAC criteria — all on the open model.

```python
!cd dev-eval-agents && python3 scripts/open_runner.py \
    --model qwen2.5:7b --num-ctx 16384 \
    --out docs/open-model-demo-output.md
```

### 5. Show the saved evaluation

```python
print(open('dev-eval-agents/docs/open-model-demo-output.md', encoding='utf-8').read())
```
